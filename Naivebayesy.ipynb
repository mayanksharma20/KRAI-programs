```python
# 8 Bayesian classification on any dataset.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
x=iris.data[:,]
#four features value of iris flower
```

```python
y=iris.target
y
#corrsponding three flower classes 0, 1,2
```

```python
print("Features=",iris['feature_names'])
```

```python
iris_dataframe = pd.DataFrame(data=np.c_[iris['data'],iris['target']], columns=iris['feature_names']+['target'])
```

```python
plt.figure()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt = pd.plotting.scatter_matrix(iris_dataframe, c=iris['target'],
                                 figsize=(15,5),
                                 s=60,alpha=1)
plt.show()
```

```python
#step-3 visualising corelation &checking assumptions of Naive bayes
import matplotlib.pyplot as plt
import seaborn as sns
dataplot = sns.heatmap(iris_dataframe.corr(), annot=True)
plt.show()
```

```python
#step-4  split dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.25 ,random_state=0 )
```

```python
#step-5  fit the model
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train, y_train)
```

```python
#step-6  Evaluate the model
import matplotlib.pyplot as plt
Y_pred = NB.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_pred)
df_cm = pd.DataFrame(cm, columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
sns.heatmap(df_cm, annot=True) #font size
plt.show()
```
